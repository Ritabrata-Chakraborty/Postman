#### Remove Outliers
Detecting Outliers:
Sorting,
Graphing,
Z-score,
Density-based techniques,
Boxplots,
Inter Quantile Range(IQR)

Treating Outliers:
Trimming,
Capping,
Replacing(Mean/Median)

```python
# Remove Outliers
# Function to Remove Outliers using the IQR Method
def remove_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    threshold = 3.5
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return column[(column >= lower_bound) & (column <= upper_bound)]

# Apply the Function to each Numerical Column in the DataFrame
numerical_columns = df.select_dtypes(include=[np.number]).columns  # Select Numerical Columns
for col in numerical_columns:
    df[col] = remove_outliers_iqr(df[col])

df.dropna(inplace=True)

# Remove Duplicates
df = df.drop_duplicates()

df
```

```python
df['DRK_YN'] = df['DRK_YN'].map({'Y': 1, 'N': 0})
df['sex'] = df['sex'].map({'Male': 1, 'Female': 0})
df
```

### Model Training
#### Target Variables: 'SMK_stat_type_cd', 'DRK_YN'

```python
# Specify the Target Variables
target_variable1 = 'SMK_stat_type_cd'
target_variable2 = 'DRK_YN'

# Split the Data into Features (X) and Target Variables (y1 and y2)
X = df.drop(columns=[target_variable1, target_variable2])
y1 = df[target_variable1]
y2 = df[target_variable2]

# Split the Data into Training and Testing Sets
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.2, random_state=42)

# Standardize the Features (Optional)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Function to Train and Evaluate a Model
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test):
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time

    start_time = time.time()
    predictions = model.predict(X_test)
    prediction_time = time.time() - start_time

    accuracy = accuracy_score(y_test, predictions)

    print(f"Model: {model}")
    print(f"Accuracy: {accuracy}")
    print(f"Training Time: {training_time} seconds")
    print(f"Prediction Time: {prediction_time} seconds")
    print(f"Memory Usage: {sys.getsizeof(model)} bytes")
    print("\n")

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Naive Bayes": MultinomialNB(),
    "Neural Network (MLP)": MLPClassifier(),
}

print("SMK_stat_type_cd")
for model_name, model in models.items():
    train_and_evaluate_model(model, X_train, y1_train, X_test, y1_test)

print("DRK_YN")
for model_name, model in models.items():
    train_and_evaluate_model(model, X_train, y2_train, X_test, y2_test)
```

### Train Models (Sequential Cross Validation)
#### Target Variables: 'SMK_stat_type_cd', 'DRK_YN'

```python
# Dependent and Independent Variables
X = df.drop(['SMK_stat_type_cd', 'DRK_YN'], axis=1)  # Drop the two target variables
y1 = df['SMK_stat_type_cd']  # First target variable
y2 = df['DRK_YN']  # Second target variable

# Function to Train and Evaluate a Model using Sequential Cross-Validation
def train_and_evaluate_model(model_name, model, X, y, task_name, n_splits):
    start_time = time.time()
    accuracy_scores = []

    for train_idx, test_idx in sequential_cross_validation(X, n_splits):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        accuracy = accuracy_score(y_test, predictions)
        accuracy_scores.append(accuracy)

    training_time = time.time() - start_time  
    
    print(f"Task: {task_name}")
    print(f"Model: {model}")
    print(f"Average Accuracy: {sum(accuracy_scores) / len(accuracy_scores)}")
    print(f"Training Time: {training_time} seconds")
    print(f"Memory Usage: {sys.getsizeof(model)} bytes")
    print("\n")

# Function for Sequential Cross-Validation
from sklearn.model_selection import TimeSeriesSplit

def sequential_cross_validation(X, n_splits):
    tscv = TimeSeriesSplit(n_splits=n_splits)
    for train_idx, test_idx in tscv.split(X):
        yield train_idx, test_idx
```
